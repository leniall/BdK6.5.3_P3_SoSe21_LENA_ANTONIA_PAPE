```python
pmids = ["31452104",
         "31437182",
         "31455877",
         "31535994"]
```

```python
str(pmids)
```

```python
for pmid in pmids:
    full_url = base_url + pmid
    pmid_json_data = urllib.request.urlopen(full_url).read()
    pmid_data = json.loads(pmid_json_data)
    print(pmid_data["result"][str(pmid)]["title"] + pmid_data["result"][str(pmid)]["pubdate"])
```

```python
for pmid in pmids:
    full_url = base_url + pmid
    pmid_json_data = urllib.request.urlopen(full_url).read()
    pmid_data = json.loads(pmid_json_data)
    text_title = pmid_data["result"][str(pmid)]["title"]
    year = pmid_data["result"][str(pmid)]["pubdate"]
    print(f"{text_title} wurde {year} publiziert.")
```
